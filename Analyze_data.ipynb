## Find the geolocation of the cities

```python
GEONAMES_SEARCH_URL = 'http://api.geonames.org/searchJSON'
def search(url, params):
    """ This function returns a json file containing the geographical results from the
    Geonames or google map websites """
    try:
        response = requests.get(url, params)
    except:
        raise ServiceException(url, params)
    return response.json()
```

```python
cities = {}
for name in name_cities:
    cities[name] = []
    params = {'q': name, 'maxRows': 1, 'country': 'CH', 'username': 'sorooshafiee'}
    response = search(GEONAMES_SEARCH_URL, params)
    cities[name].append(response['geonames'][0]['lat'])
    cities[name].append(response['geonames'][0]['lng'])
    cities[name].append(response['geonames'][0]['adminCode1'])
#print('Number of the recovered cities: {}/{}'.format(sum(x != 0 for x in lst), len(lst)))
```

```python
coordinates_cities = np.asarray(list(cities.values()))[:,:2].astype(float)
city_name = list(cities.keys())
```

## Find the cities of the geolocations based on the name of cities

```python
df['canton'] = 'Nan'
df['city'] = 'Nan'
```

```python
df.head()
```

```python
from scipy.spatial.distance import cdist
dist = np.zeros(len(coordinates_cities))
for index, row in df.iterrows():
    flag_error = 0
    lat = row['placeLatitude']
    lang = row['placeLongitude']
    geocor = np.matrix(np.asarray([lat,lang]))
    for i in range(len(coordinates_cities)):
        try:
            dist[i] = cdist(geocor,np.matrix(coordinates_cities[i]))
        except:
            flag_error = 1
            break
    if flag_error == 0:
        ind = np.argmin(dist)
        val_min = np.min(dist)
        df.set_value(index,'city',city_name[ind])
        df.set_value(index,'canton',cities[city_name[ind]][2])
```
