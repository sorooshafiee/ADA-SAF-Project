```python
def geocoding_of_region(url,parameters):
    response = requests.get(url,params)
    loc=response.json()['results'][0]['geometry']['location']
    lat=loc['lat']
    lng=loc['lng']
    return lat,lng
```

```python
latlngs=[]
url='https://maps.googleapis.com/maps/api/geocode/json'
for reg in uniques:
    params={'address':reg,'key':api_key}
    res=geocoding_of_region(url,params)
    latlngs.append(res)
```

```python
all_unique_paths=([list(i) for i in set(map(tuple, moves))])
```

```python
def conv_to_coordinates(path):
    p1=path[0]
    p2=path[1]
    #depart point:
    c=0
    points=[]
    for i in uniques:
        if p1==i:
            k=latlngs[c]
            points.append(k)
            break
        c+=1;
    #arrive point:
    c=0
    for i in uniques:
        if p2==i:
            k=latlngs[c]
            points.append(k)
            break
        c+=1;
            
    return points
         
```
